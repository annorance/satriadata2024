```python
# Function to clean usernames from a tweet
def clean_tweet(tweet):
    # Define regex pattern to match usernames and hyperlinks
    pattern = r'(@[\w/+=]+)|(https?://\S+)'

    # Remove usernames and hyperlinks
    tweet = re.sub(pattern, '', tweet)

    return tweet

# Apply the cleaning function to every row in the 'tweet' column
df['text'] = df['text'].apply(clean_tweet)
print(df)
```

```python
#Convert text to lowercase
#buat fungsi
def lowercase(text):
  text = text.lower()
  return text

#apply fungsi ke data
df['text'] = df['text'].apply(lowercase)
print(df)
```

```python
#Remove special characters using regular expressions (re)
def special_char(text):
  text = re.sub(r'[^\w\s]', '', text)
  return text

#apply fungsi ke data
df['text'] = df['text'].apply(special_char)
print(df)
```

```python
#Menghilangkan emoji --> ERROR MULU AAAAAAAAAAAAAAAA
# Contoh DataFrame dengan teks yang memiliki karakter terdistorsi
data = {
    'text': [
        "Terdapat beberapa emoji dalam teks ini: ÃƒÂ°Ã‚Å¸Ã‚Â¤Ã‚Â dan juga Ã°Å¸Ëœâ€¢",
        "Ini adalah contoh teks tanpa emoji.",
        "Contoh lain dengan emoji: ÃƒÂ°Ã‚Å¸Ã‚Â¤Ã‚Â"
    ]
}
df = pd.DataFrame(data)

# Fungsi untuk mencoba mengkonversi teks terdistorsi kembali ke bentuk asli
def decode_text(text):
    try:
        return text.encode('latin1').decode('utf-8')
    except (UnicodeEncodeError, UnicodeDecodeError):
        return text

# Terapkan konversi ke kolom teks
df['decoded_text'] = df['text'].apply(decode_text)

# Pola regex untuk mendeteksi emoji
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # Emoticons
                           u"\U0001F300-\U0001F5FF"  # Simbol & misc
                           u"\U0001F680-\U0001F6FF"  # Transportasi & map
                           u"\U0001F1E0-\U0001F1FF"  # Flag (iOS)
                           "]+", flags=re.UNICODE)

# Fungsi untuk mendeteksi emoji dalam teks
def detect_emoji(text):
    if isinstance(text, str):
        detected_emoji = emoji_pattern.findall(text)
        return detected_emoji
    else:
        return []

# Terapkan fungsi detect_emoji pada setiap elemen dalam kolom 'decoded_text'
df['detected_emojis'] = df['decoded_text'].apply(detect_emoji)

# Tampilkan hasil
print(df[['text', 'decoded_text', 'detected_emojis']])
```
